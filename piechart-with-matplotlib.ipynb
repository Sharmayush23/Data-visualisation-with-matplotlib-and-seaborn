```python
# make 2 variable for labels and values 
labels = newdata.PaymentMethodCategory
values = newdata.Count
type(values)
```

# Visualization 

```python
# Simple Pie Chart
plt.figure(figsize=(5,4))
# Create a pieplot
plt.pie(values)
plt.show()
```

# Add labels

```python
# Label distance: gives the space between labels and the center of the pie
plt.figure(figsize=(5,4))
plt.pie(values, labels=labels, labeldistance=0.5);
plt.show();
```

# Add title

```python
plt.figure(figsize=(5,4))
# Label distance: gives the space between labels and the center of the pie
plt.pie(values, labels=labels, labeldistance=1.15);
plt.title("Pie Chart", size=15)
plt.show();
```

# Customize wedges
- customize wedges to get some blank spaces around them

```python
plt.figure(figsize=(5,4))
color = ['lightcoral', 'rosybrown']
plt.pie(values, labels=labels, labeldistance=1.15, wedgeprops = { 'linewidth' : 2, 'edgecolor' : 'white' },colors=color)
plt.title("Customize wedges", size=15)
plt.show()
```

#  Custom Color

```python

color = ['lightcoral', 'rosybrown', 'sandybrown', 'navajowhite', 'gold',
        'khaki', 'lightskyblue', 'turquoise', 'lightslategrey', 'thistle', 'pink']
plt.figure(figsize=(5,4))
plt.title("Custom Color", size=15)
plt.pie(values, labels=labels, labeldistance=1.15, wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, colors=color);
plt.show()
```

# Auto-label slices
- Pass a function or format string to autopct(auto percentages) to label slices.


```python
plt.figure(figsize=(5,4))
plt.pie(values, labels=labels, 
        labeldistance=1.15, 
        wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
        autopct='%2.2f%%');
plt.title("Custom Color", size=15)
plt.show();
```

# Swap label and autopct text positions

```python
plt.figure(figsize=(5,4))
plt.pie(values, labels=labels, 
        wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
        autopct='%1.1f%%',
        pctdistance=1.25, 
        labeldistance=.6);
plt.show();
```

# Explode, shade, and rotate slices

```python
plt.figure(figsize=(5,4))
explode = (0, 0, 0.1, 0)  # only "explode" the Electronic check
plt.pie(values, labels=labels, 
        explode=explode,
        labeldistance=1.15, 
        wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }, 
        autopct='%1.1f%%',
        shadow=True, startangle=45)
plt.show()
```

# Add legend

```python
plt.figure(figsize=(10,10))

plt.pie(values, 
        wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }, 
        autopct='%1.1f%%',)
plt.legend(labels,loc="best", fontsize=12)

plt.show()
```

# Legend  Header add


```python
plt.figure(figsize=(18,10))

plt.pie(values, 
        wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }, 
        autopct='%1.1f%%',)
plt.legend(labels, loc='best', fontsize=12,title = "Category")

plt.show()
```

# Double Pie Chart 
- For this you need to groupby the data  2 times. First group 2 category variable and next group only main variable 

```python
# here groupby Payment Method and make a new data 
newdata2 = data.groupby(['PaymentMethod','gender']).size().reset_index(name='Count')
newdata2.head(10)
```

```python
newdata3 = data.groupby('PaymentMethod').size().reset_index(name='Count')
newdata3.head(10)
```

```python
#Plot
colors = ['#ff6666', '#ffcc99', '#99ff99', '#66b3ff']
colors_gender = ['g','y', 'g','y','g','y','g','y',]
plt.pie(newdata3.Count, labels=newdata3.PaymentMethod, colors=colors, startangle=90,frame=True,radius=1,autopct='%1.1f%%',
        pctdistance=0.85,)
plt.pie(newdata2.Count,colors=colors_gender,autopct='%1.1f%%',
        startangle=90, radius=.7 )
centre_circle = plt.Circle((0,0),.2,color='black', fc='white',linewidth=0,)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.text(2, 1, "Green = Female \nYellow = Male ", bbox=dict(facecolor='white', alpha=0.5))
plt.show()
```

#  Make a function  for showing

```python
def plot_pie_charts(x, y, title):
    # more muted color 
    c = ['lightcoral', 'rosybrown', 'sandybrown', 'navajowhite', 'gold',
        'khaki', 'lightskyblue', 'turquoise', 'lightslategrey', 'thistle', 'pink']
    plt.figure(figsize=(9,9))
    plt.title(title, size=20)
    plt.pie(x, labels=y, labeldistance=1.15, wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, colors=c,autopct='%1.1f%%');
    plt.legend(y, loc='best', fontsize=12)
    plt.show()
```

# First Chart using function

```python
plot_pie_charts(values, labels,"Payment Method")
```
