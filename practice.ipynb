```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import csv
from collections import Counter

plt.style.use("fivethirtyeight")
ages=[23,22,25,26,27,22,28,29,30]

x_value=np.arange(len(ages))

py_dev=[2323,243,9432,324,342,567,4675,123,843]

width=.25
plt.bar(x_value+width,py_dev,width=width,color="black",label="py developer")

js_dev=[234,345,678,908,678,345,565,678,111]

plt.bar(x_value,js_dev,width=width,color ="orange",label="js developer")
c_dev=[345,676,887,989,333,456,654,564,444]

plt.bar(x_value-width,c_dev,width=width,color="blue",label="c developer")
plt.legend()
plt.xlabel("ages")
plt.ylabel("no of the dev")

plt.xticks(ticks=x_value,labels=ages)
plt.tight_layout()

plt.show()

```
