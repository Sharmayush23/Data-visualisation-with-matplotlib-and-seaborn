```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import csv
from collections import Counter



plt.style.use("fivethirtyeight")

with open("ds_salaries.csv") as csv_file:
    #csv_reader is the container what store the dic data typw values coming from the csv.Dictreader 
    csv_reader = csv.DictReader(csv_file)

    row = next(csv_reader)
    print(row)
    
    
```

```python
print(row['job_title'].split(';'))
```

```python
with open("ds_salaries.csv") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    job_rule=Counter()
    for row in csv_reader:
        job_rule.update(row["job_title"].split(";"))
    

job_title=[]
popularity=[]

for item in job_rule.most_common(15):
    job_title.append(item[0])
    popularity.append(item[1])

print(job_title)
print(popularity)
```

```python
job_title.reverse()
popularity.reverse()
plt.barh(job_title,popularity,)
plt.title("most popular job title")

#plt.xlabel("job titles ")
#plt.ylabel("no of people work")
plt.tight_layout()

plt.show()
```

```python
with open("punjab-crime-stats.csv") as csv_file:
    pb_data=csv.DictReader(csv_file)
     
    pb_crime=Counter()

    for item in pb_data:
        pb_crime.update(item["CrimeType"].split(";"))

print(pb_crime.most_common(15))        

```

```python
crime_type=[]
cases=[]

for item in pb_crime.most_common(15):
    crime_type.append(item[0])
    cases.append(item[1])

print(crime_type)
print(cases)
```

```python
plt.barh(crime_type,cases,color='orange')
```
