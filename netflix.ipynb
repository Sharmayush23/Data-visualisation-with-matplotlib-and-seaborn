```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import csv
from collections import Counter

df=pd.read_csv("most-popular.csv")
df.head()
```

```python
most_popular=df["show_title"]
most_popular
```

```python
from requests import Response


most=Counter()

for i in most_popular:
    most.update(i.split(';'))

most_famous=[]
popularity=[]

for item in most.most_common(15):
    most_famous.append(item[0])
    popularity.append(item[1])
most_famous.reverse()
popularity.reverse()

plt.barh(most_famous, popularity)

plt.title("Most Popular ")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()
plt

plt.show()
```
