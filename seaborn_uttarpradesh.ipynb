```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Category': ['Category A', 'Category B', 'Category C', 'Category D'],
    'Value': [30, 50, 20, 40]
}

# Create a DataFrame
import pandas as pd
df = pd.DataFrame(data)

# Create a FacetGrid
g = sns.FacetGrid(df, col="Category", col_wrap=2, height=4, margin_titles=True)

# Create a horizontal bar plot using FacetGrid
g.map(sns.barplot, "Value", "Category", color='skyblue')

# Customize ticks using Matplotlib functions
for ax in g.axes.flat:
    ax.set_xticks(range(0, 60, 10))
    ax.set_xticklabels(['0', '10', '20', '30', '40', '50'])

# Adjust other settings if needed
g.set_axis_labels('Value', 'Category')
g.set_titles(col_template="{col_name} Category")
g.fig.suptitle('Horizontal Bar Chart', y=1.02)

# Show the plot
plt.show()
```
