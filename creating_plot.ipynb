```python
from matplotlib import pyplot as plt
import numpy as np

import matplotlib as mpl

print(plt.style.available)
plt.style.use("fivethirtyeight")

maths_score=[7,69,90,90,95,47,57,76,78]
science_score=[72,69,90,90,95,47,57,76,56]

plt.plot(maths_score,science_score)

plt.title("simple graph")
plt.xlabel("maths score")
plt.ylabel("science score")
plt.legend()
plt.show()
```

```python
x=np.arange(3,10)
x
plt.style.use("fivethirtyeight")
plt.figure(figsize=(5,2.7))
plt.plot(x,x**2,color="black",linewidth=2,linestyle='--')
plt.plot(x,x**3,"v")
plt.xlabel("x ordinate",color="black")
plt.ylabel("y coordinate")

plt.title("simple graph")
plt.legend(["x","x sqauare"])
plt.grid()
plt.show()
```

```python
x=np.arange(3,10)
x
#style of the graph
plt.style.use("ggplot")
plt.figure(figsize=(5,2.7))
plt.plot(x,x**3,color="black",linewidth=2,linestyle='--')
plt.plot(x,1/x,"v")
plt.xlabel("x ordiplote",color="black")
plt.ylabel("y coordinate")

plt.title("simple graph")
plt.legend(["x","1/x66"])
plt.grid()
plt.savefig("simple.png")
plt.tight_layout()
plt.show()
```
