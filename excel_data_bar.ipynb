```python


  # Read the Excel file into a DataFrame
file_path = "Uttar_Pradesh_Dataset.xlsx"
df = pd.read_excel(file_path)

# Create a dictionary to store plot type counts
plot_type_counts = {}

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Split plot types and update the dictionary
    plot_types = row["Type_of_plot"].split(";")
    for plot_type in plot_types:
        plot_type = plot_type.strip()  # Remove leading/trailing spaces
        plot_type_counts[plot_type] = plot_type_counts.get(plot_type, 0) + 1

# Plot the bar graph
plt.bar(plot_type_counts.keys(), plot_type_counts.values())
plt.xlabel("Plot Type")
plt.ylabel("Count")
plt.title("Plot Type Distribution")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better visibility
plt.show()
```
